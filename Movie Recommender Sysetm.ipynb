# Importing Modules

```python
import numpy as np
import pandas as pd
import ast
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
movies_df=movies.copy()
```

# Merging movies and credits Datasets

```python
movies_df=movies_df.merge(credits,on='title')
movies_df.head()
```

```python
movies_df.info()
```

# Feature Selection

```python
movies_df= movies_df[['id','genres','keywords','overview','title','cast','crew']]
movies_df.head()
```

# Data Preprocessing

```python
movies_df.isna().sum()
```

```python
movies_df.dropna(inplace=True)
```

```python
movies_df.duplicated().sum()
```

```python
# Function to handel genres
def convert(text):
    genres=[]
    for i in ast.literal_eval(text):
        genres.append(i['name'])
    return genres
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
```

```python
movies_df.head()
```

```python
movies_df['keywords']=movies_df['keywords'].apply(convert)
```

```python
movies_df.head()
```

```python
def FetchCast(text):
    
    cast=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter!=5:
            cast.append(i['character'])
            counter+=1
            break
    return cast
```

```python
movies_df['cast']=movies_df['cast'].apply(FetchCast)
```

```python
movies_df.head()
```

```python
def FetchDirector(text):
    
    director=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            director.append(i['name'])
            break
    return director
```

```python
movies_df['crew']=movies_df['crew'].apply(FetchDirector)
```

```python
movies_df.head()
```

```python
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df.head()
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies_df['keywords']=movies_df['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies_df['cast']=movies_df['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies_df['crew']=movies_df['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags']= movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
```

```python
movies_df.head()
```

```python
new_df=movies_df.copy()
```

```python
new_df=new_df[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

# Vectorization of text to numbers

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()[0:1000]
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

# Checking Similarity of movies

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def Recommender(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
         print(new_df.iloc[i[0]].title)
```

```python
Recommender('Batman')
```
